```python
import numpy as np
import tensorflow as tf
import pandas as pd
```

# Load the data

```python
data = pd.read_csv(r"C:\Users\13177\Downloads\poisson_regression_data-1 (1).csv")
input_data = data.iloc[:, 1:].values
target_values = data.iloc[:, 0].values
```

# Prepare the inputs

```python
target_values = target_values.reshape(-1, 1)
ones = np.ones((input_data.shape[0], 1))
inputs = np.concatenate([ones, input_data], axis=1)
```

# Initialize the weights

```python
tf.random.set_seed(100)
weights = tf.Variable(tf.random.normal([6, 1]))
```

# Set the training parameters

```python
training_epochs = 1000
learning_rate = 0.01
```

# Convert the inputs to tensors

```python
inputs = tf.cast(inputs, dtype=tf.float32)
target_values = tf.cast(target_values, dtype=tf.float32)
```

# Define the loss function

```python
def poisson_regression_loss(y_true, y_predicted):
    return tf.reduce_mean(y_predicted - y_true * tf.math.log(y_predicted))
```

# Train the model

```python
optimizer = tf.optimizers.SGD(learning_rate)
for epoch in range(training_epochs):
    with tf.GradientTape() as tape:
        predicted_values = tf.exp(tf.matmul(inputs, weights))
        loss = poisson_regression_loss(target_values, predicted_values)
    gradients = tape.gradient(loss, weights)
    optimizer.apply_gradients(zip([gradients], [weights]))
```
